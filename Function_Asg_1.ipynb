## **Function Assignment**

```python
# 1) Sum of all even numbers.

def sum_even(numbers):
    return sum(num for num in numbers if num % 2 == 0)

print(sum_even([1,2,3,4,5,6]))
```

```python
# 2) Reverse String .

def reverse_string(s):
    return s[::-1]

print(reverse_string('India is my country'))
```

```python
# 3) Squares of Integers.

def sqr_num(lst):
    return [x ** 2 for x in lst]

sqr_num([3, 4, 5, 2, 6])
```

```python
# 4) Check Prime Numbers(1, 200).

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5)+ 1):
        if n % i == 0:
            return False
    return True

for number in range(1, 200):
    if is_prime(number):
        print(number, end = ' ')
```
