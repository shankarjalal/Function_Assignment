## **Function Assignment**

```python
# 1) Sum of all even numbers.

def sum_even(numbers):
    return sum(num for num in numbers if num % 2 == 0)

print(sum_even([1,2,3,4,5,6]))
```

```python
# 2) Reverse String .

def reverse_string(s):
    return s[::-1]

print(reverse_string('India is my country'))
```

```python
# 3) Squares of Integers.

def sqr_num(lst):
    return [x ** 2 for x in lst]

sqr_num([3, 4, 5, 2, 6])
```

```python
# 4) Check Prime Numbers(1, 200).

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5)+ 1):
        if n % i == 0:
            return False
    return True

for number in range(1, 200):
    if is_prime(number):
        print(number, end = ' ')
```

```python
#5) Iterator of Fibonacci numbers.

def fibonacci_iterator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib_iter = fibonacci_iterator()
for _ in range(10):             # example using
    print(next(fib_iter), end=' ') 
```

```python
#6) Generator for pawer of 2

def power_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for val in power_of_two(10):
    print(val, end=' ')
```

```python
#7) Generator that yields file lines

def read_file_line_by_line(file):
    with open(file, 'r') as f:
        for line in f:
            yield line.strip
```

```python
#8) Sort list of tuples by second elements(lambda function).

data = [(1, 'apple'), (3, 'banana'), (2, 'cherry')]
sort_data = sorted(data, key = lambda x:x[1])
print(sort_data)
```

```python
#9) Convert Celsius to Fahrenheit(map function)

temp_celsius = [0, 20, 38, 100]
temp_fahrenheit = list(map(lambda c: (c * 9/5)+32, temp_celsius))
print(f'Temperature in Fahrenheit: {temp_fahrenheit}')
```

```python
#10) Remove Vowrels usin filter().

def remove_vowels(s):
    return ''.join(filter(lambda char: char.lower() not in 'aeiouAEIOU', s))

print(remove_vowels('This is an eample of string.'))
```
